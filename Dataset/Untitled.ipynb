```python
import json

# If reading from a file
with open('./preference_data_clean.json', 'r') as file:
    data = json.load(file)

# Count objects based on structure
if isinstance(data, list):
    count = len(data)
    print(f"Number of objects in array: {count}")
elif isinstance(data, dict):
    count = len(data)
    print(f"Number of key-value pairs: {count}")
else:
    print("Not a standard JSON structure")
```
